```python
# Example: Counting number of tracks in each genre
genre_counts = df['track_genre'].value_counts()
print("\nNumber of tracks in each genre:")
print(genre_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['duration_ms'] / 60000, bins=30, kde=True, color='skyblue')
plt.title('Distribution of Track Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot: Average popularity by genre
avg_popularity = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False)

plt.figure(figsize=(20, 10))
sns.barplot(x=avg_popularity.index, y=avg_popularity.values, palette='coolwarm')
plt.title('Average Popularity by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Box plot: Loudness distribution by mode
plt.figure(figsize=(8, 6))
sns.boxplot(x='mode', y='loudness', data=df, palette='Set2')
plt.title('Loudness Distribution by Mode')
plt.xlabel('Mode (Major=1, Minor=0)')
plt.ylabel('Loudness (dB)')
plt.xticks([0, 1], ['Minor', 'Major'])
plt.show()
```

```python
# Selecting numeric columns for pair plot
numeric_columns = ['popularity', 'duration_ms', 'danceability', 'energy', 'loudness', 'tempo']

# Creating pair plot
sns.pairplot(df[numeric_columns])
plt.suptitle('Pairwise Relationships in Spotify Tracks Dataset', y=1.02)
plt.show()
```

```python
# Calculate the average duration in milliseconds
avg_duration_ms = df['duration_ms'].mean()

print(f'Average Duration (ms): {avg_duration_ms:.2f}')
```

```python
max_duration_ms = df['duration_ms'].max()

print(f'Max Duration (ms): {max_duration_ms:.2f}')
```

```python
min_duration_ms = df['duration_ms'].min()

print(f'Min Duration (ms): {min_duration_ms:.2f}')
```

```python
# Plotting a histogram of track popularity
plt.figure(figsize=(10, 6))
plt.hist(df['popularity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plotting a box plot of energy by genre
plt.figure(figsize=(20, 8))
sns.boxplot(x='track_genre', y='energy', data=df, palette='Set2')
plt.title('Energy Levels by Genre')
plt.xlabel('Genre')
plt.ylabel('Energy')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting a pie chart
# Count tracks by mode (0 = minor, 1 = major)
mode_counts = df['mode'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(mode_counts, labels=['Minor', 'Major'], autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightblue'])
plt.title('Distribution of Tracks by Mode')
plt.axis('equal')
plt.show()
```

```python
# Select numerical columns for correlation analysis
cols_to_correlate = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

# Calculate correlation matrix
corr_matrix = df[cols_to_correlate].corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of Audio Features')
plt.show()
```
